```python
import pandas as pd

#name.basics.tsv
name_basics=pd.read_csv('name.basics.tsv',delimiter='\t')
print(list(name_basics.columns.values)) #file header
print(name_basics.head(5))

#title.ratings.tsv
ratings=pd.read_csv('title.ratings.tsv',delimiter='\t')
print(list(ratings.columns.values)) #file header
print(ratings.head(5))

#title.principals.tsv
principals=pd.read_csv('title.principals.tsv',delimiter='\t')
print(list(principals.columns.values)) #file header
print(principals.head(5))

#title.basics.tsv
title_basics=pd.read_csv('title.basics.tsv',delimiter='\t')
print(list(title_basics.columns.values)) #file header
print(title_basics.head(5))

#title.akas.tsv
akas=pd.read_csv('title.akas.tsv',delimiter='\t')
print(list(akas.columns.values)) #file header
print(akas.head(5))
```

```python
akas.rename(columns={'titleId':'tconst'}, inplace=True)
akas.head(3)

merge1 = pd.merge(title_basics,ratings,
                 on='tconst')
merge2 = pd.merge(merge1,principals,
                 on='tconst')
merge3 = pd.merge(merge2,title_basics,
                 on='tconst')
title = pd.merge(merge3,akas,
                 on='tconst')

title.head()
```

```python
names = name_basics.copy()

def missing_values_table(df):
        mis_val = df.isnull().sum()
        mis_val_percent = 100 * df.isnull().sum() / len(df)
        mis_val_table = pd.concat([mis_val, mis_val_percent], axis=1)
        mis_val_table_ren_columns = mis_val_table.rename(
        columns = {0 : 'Missing Values', 1 : '% of Total Values'})
        mis_val_table_ren_columns = mis_val_table_ren_columns[
            mis_val_table_ren_columns.iloc[:,1] != 0].sort_values(
        '% of Total Values', ascending=False).round(1)
        print ("Your selected dataframe has " + str(df.shape[1]) + " columns.\n"      
            "There are " + str(mis_val_table_ren_columns.shape[0]) +
              " columns that have missing values.")
        return mis_val_table_ren_columns

missing_values_table(names)
```
